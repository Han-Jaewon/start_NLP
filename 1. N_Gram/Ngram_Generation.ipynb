# ngram 함수 정의

```python
def myngram(txt, cw, n):
    
    '''
    txt: 문자열
    cw: char or word
    n: 1, 2, 3, ... (uni, bi, tri ...)
    '''
    
    txt = re.sub(r'(?<=\w)[,.](?=\s|$)', '', txt)
    txt = re.sub(r"'", "", txt)
    words = txt.split()
    dic = {}
    
    if cw == 'char':
        for i in range(len(txt)-n+1):
            ngram = txt[i:i+n]
            ngram = "".join(ngram)
            if ngram not in dic:
                dic[ngram] = 0 
            else:
                dic[ngram] += 1
    elif cw == 'word':
        for i in range(len(words)-n+1):
            ngram = words[i:i+n]
            ngram = " ".join(ngram)
            if ngram not in dic:
                dic[ngram] = 0 
            else:
                dic[ngram] += 1

    
    result = sorted(dic.items(), key = lambda x: x[1], reverse = True)
    
    for key, value in result:
        print(value, key, sep='\t')
```
