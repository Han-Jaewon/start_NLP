```python
import re
from collections import defaultdict
```

```python
def myngram2(file, find_word, n):
    
    '''
    file: 입력값. 엔그램으로 만들 대상 파일
    find_word: 찾을 문자열. ex) '학교에'
    n: 엔그램 수. ex) 1, 2, 3, ...
    '''
    
    ngram = defaultdict(int)
    result = defaultdict(int)
    
    with open(file, encoding='utf-8') as f:
        for line in f:
            line_ = re.sub("(?=[,.])", "", line)
            words = line_.split()
            
            # ngram 생성
            for i in range(n):
                for j in range(len(words)-i):
                    ngram_ = tuple(words[j:j+i+1])
                    ngram[ngram_] += 1
                    
        for k, v in sorted(ngram.items(), key = lambda x:x[1], reverse=True):
            if k[0] == find_word:
                result[k] += v
                
        for w, c in result.items():
            t = sum(result.values())
            print(c/t, w, sep = '\t')
```

```python
myngram2('ngram_test_corpus.txt', '학교에', 3)
```
